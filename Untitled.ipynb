## Mask Board

```python
masked = image.copy()
masked[1080:2430, 1120:3350,:] = [0, 0, 0]

plt.imshow(masked)
```

## Color Segmentation

```python
blue = cv2.inRange(image, (40, 100, 150), (70, 160, 250))
red = cv2.inRange(image, (180, 0, 0), (255, 95, 95))

blue_segmented = cv2.bitwise_and(image, image, mask = red)
red_segmented = cv2.bitwise_and(image, image, mask = blue)

segmented = cv2.bitwise_or(blue_segmented, red_segmented)

# segmented = cv2.bitwise_and(image, image, mask = red)
# segmented = cv2.bitwise_and(image, image, mask = blue)
plt.imshow(segmented)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
kernel = np.ones((10, 10), np.uint8)
dilated = cv2.dilate(segmented.copy(), kernel, iterations = 3)
eroded = cv2.erode(dilated.copy(), kernel, iterations = 5)

blurred = cv2.medianBlur(eroded, 7)
plt.imshow(eroded)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
gray = cv2.cvtColor(blurred.copy(), cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = "gray")
```
